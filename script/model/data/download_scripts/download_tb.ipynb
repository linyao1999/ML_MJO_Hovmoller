```python
import os
import requests
from datetime import datetime, timedelta

# --------------------------
# USER SETTINGS
# --------------------------
start_year = 1999
end_year = 1999
outdir = "/pscratch/sd/l/linyaoly/MJO_ML_2025/script/model/data/TB"

os.makedirs(outdir, exist_ok=True)

# --------------------------
# Loop over time
# --------------------------
start_date = datetime(start_year, 6, 19)
end_date = datetime(end_year, 6, 19)

# hours = ["00", "03", "06", "09", "12", "15", "18", "21"]
hours = ["03"]  # only download 03UTC files for 8x daily data   

t = start_date
while t <= end_date:
    yyyy = t.strftime("%Y")
    mm = t.strftime("%m")
    dd = t.strftime("%d")
    yyyymmdd = t.strftime("%Y.%m.%d")

    for hh in hours:
        # filename = f"GRIDSAT-B1.{yyyy}.{mm}.{dd}.{hh}.v02r01.nc"
        filename = 'GRIDSAT-B1.1999.06.19.03.v02r01.nc'
        url = f"https://www.ncei.noaa.gov/thredds/fileServer/cdr/gridsat/{yyyy}/{filename}"
        outfile = os.path.join(outdir, filename)

        if os.path.exists(outfile):
            os.remove(outfile)
            print("Already exists:", filename)
            # continue

        print("Downloading:", url)
        r = requests.get(url, stream=True)

        if r.status_code == 200:
            with open(outfile, "wb") as f:
                for chunk in r.iter_content(chunk_size=8192):
                    f.write(chunk)
        else:
            print("Missing file:", url)

    t += timedelta(days=1)

print("DONE")
```
