```python
import os
import optuna
from optuna.trial import TrialState
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
for dataflg in ["raw", "mjo", "rossby", "kelvin", "gravity"]:
    study_name = f'UNet_A_lat10_lead35_in_olr_tar_ROMI_mem95_{dataflg}'  
    storage_name = f"sqlite:////pscratch/sd/l/linyaoly/MJO_ML_2025/script/model/exp/UNet_hov_to_leads/optuna/{study_name}.db"
    db_path = storage_name[10:]

    if os.path.exists(db_path):
        study = optuna.load_study(study_name=study_name, storage=storage_name)
        completed_trials = [t for t in study.trials if t.state == TrialState.COMPLETE]

        # Extract trial numbers and values (i.e., losses)
        trial_numbers = [t.number for t in completed_trials]
        losses = [t.value for t in completed_trials]

        # Plot

        plt.plot(trial_numbers, losses, marker='o', linestyle='-', label=dataflg)
        plt.xlabel('Trial Number')
        plt.ylabel('Loss')
        # plt.title(f'Loss per Trial - {dataflg}')
        plt.grid(True)
        plt.tight_layout()

    else:
        print(f"{storage_name} does not exist")

plt.legend()
```
