```python
import optuna 
import numpy as np 
import os 
```

```python
fn = './CNN_MLP_lat20_lead35_in_olr_tar_ROMI_fltano120_mem95.db'
study = optuna.load_study(study_name='CNN_MLP_lat20_lead35_in_olr_tar_ROMI_fltano120_mem95', storage=f'sqlite:///{fn}')
print(f'Number of trials: {len(study.trials)}')
```

```python
study.best_trial
print(f'Best trial: {study.best_trial.number}')
```

```python
from optuna.trial import TrialState
import matplotlib.pyplot as plt

completed_trials = [t for t in study.trials if t.state == TrialState.COMPLETE]

# Extract trial numbers and values (i.e., losses)
trial_numbers = [t.number for t in completed_trials]
losses = [t.value for t in completed_trials]

# Plot

plt.plot(trial_numbers, losses, marker='o', linestyle='-')
plt.xlabel('Trial Number')
plt.ylabel('Loss')
# plt.title(f'Loss per Trial - {dataflg}')
plt.grid(True)
plt.tight_layout()
```
